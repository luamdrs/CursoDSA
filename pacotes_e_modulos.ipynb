**Podemos trabalhar com mÃ³dulos dos pacotes (quando disponÃ­veis)**

```python
# importando o modulo request do pacote urllib, usado para trazer url's para dentro do nosso ambiente Python
import urllib.request
```

```python
# variavel resposta armazena o objeto de conexao a url passada como parametro
resposta = urllib.request.urlopen('https://www.python.org/')
```

```python
# objeto resposta
print(resposta)
```

```python
# chamando o metodo read() do objeto resposta e armazenando o codigo html na variavel html
html = resposta.read()
```

```python
# imprimindo html
print(html)
```
